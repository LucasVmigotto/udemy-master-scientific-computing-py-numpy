# 06 Exercise - Working with Monochromatic Images

One common application of 2D arrays is with monochromatic images. A monochromatic image is an image with a single color channel. In contrast with color images, which has three color channels, a monocromatic image has only one channel that records the intensity of a given color. This makes monochromatic images ideal to model with 2D arrays. The two axes of the 2D array then represents the width and height of the image. The value at row $i$ and column $j$ represent the intensity of the given color (ranging from 0 to 1) at the pixel $(i, j)$.

The cell below to import the necessary libraries.

```python
# RUN THE CELL AND DON'T MODIFY IT
import numpy as np # Imports NumPy
from skimage import data # Getting the image
import matplotlib.pyplot as plt # Showing the image
```

The package <b>skimage</b> is a package for image processing in Python. We will only use the submodule <b>data</b> to get a grayscale image.  

We can now load a premade image with the <b>data.camera()</b> method and display it with the command <b>plt.imshow()</b>.

```python
# RUN THE CELL AND DON'T MODIFY IT
camera = data.camera()
plt.imshow(camera)

plt.show()
```

## Exercise 1:  Checking Out the Image

Check that the image stored in the variable <b>camera</b> indeed is a NumPy array. Print out the shape of the NumPy array.

```python
# Check that the image is a NumPy array
print(type(camera))

# Find the shape of the NumPy array
print(camera.shape)
```

You can see that the image is just a NumPy array. That means that we can apply everything we have learned about NumPy to such images.

## Exercise 2: Inspecting the Values in the Image

Print the maximum and minimum value in the image.

```python
# Print the maximum value in the image
print('Max value', camera.max())

# Print the minimum value in the image
print('Min value', camera.min())
```

Hence each entry in the image (NumPy array) is a value between 0 and 255. This value represents the intensity of the color at the given pixel.

## Exercise 3: Cropping the Image

The image is a perfect square with 512x512 pixels. Let us crop out the center of the image with dimensions 300x300.

```python
# Find how much to crop on each side. Store this in a variable called crop.

# NOTE: The crop variable you made must be turned into an integer: In the slicing below,
# only integers are valid. You can use the built in function int().
PX_CROP_PER_SIDE = int((512 - 300) / 2)

# Crop out the center of the image.
camera_300x300 = camera[PX_CROP_PER_SIDE:-PX_CROP_PER_SIDE, PX_CROP_PER_SIDE:-PX_CROP_PER_SIDE]

# Show the image with the plt.imshow function
plt.imshow(camera_300x300)
plt.show()
```

## Exercise 4: Removing Spesific Rows

Let us turn the rows 40-50 and 60-70 black in the image. To do this, we will use indexing by passing in lists.

```python
# Make a list representing the rows that should be removed
columns_2_remove = np.arange(40, 71)

# Make a variable new_copy that is a copy of camera with the .copy() function
new_copy = camera.copy()

# Remove the rows created above from the new_copy image
new_copy[columns_2_remove] = 0

# Print out the new_copy image
plt.imshow(new_copy)
plt.show()
```

## Exercise 5: Change the Intensity of the Image

We can use Boolean indexing to change the intensity of the image. As an extreme case, we can set everything with intensity over 100 to the max (which is 255).

```python
# Make a new copy of the camera image called new_copy with NumPy's copy method
new_copy = camera.copy()

# Use Boolean indexing to set everything over 100 to the max in new_copy.
new_copy[new_copy > 100] = new_copy.max()

# Show the image
plt.imshow(new_copy)
plt.show()
```

## Exercise 6: Challenge

Inverting an image means to reflect the value around the middle intensity. To do this, we can use the formula $$\textrm{reflected value} = 255 - \textrm{old value}.$$

As an example, if an intensity value is 100, then the reflected value is $$255 - 100 = 155.$$

```python
# Invert the camera image.
camera_invert = 255 - camera

# Show the image
plt.imshow(camera_invert)
plt.show()
```
